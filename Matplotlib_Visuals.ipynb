```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline

#import dataset 
df1 = pd.read_csv("/Users/elizabethdombowsky/Desktop/Book1.csv", delimiter=",")
```

```python
df1
```

# Plot of Gas Prices Over Time by Country

```python
# Create Plot and Specify color
df1.plot(x="Year", y=["USA", "UK", "Canada", "Australia", "France", "Germany", "Italy", "Japan", "Mexico", 
                      "South Korea"],kind="line", figsize=(10, 10), color = ["purple", "blue", "grey","grey", "grey", "blue","grey", "grey", "purple", "grey"])

# Create Label and Change Font size
plt.title("Gas Price Growth Over Time", fontdict= {'fontsize':30})

# Xlabel with custom font size
plt.xlabel("Year",fontdict= {'fontsize':20})

#y label with custom font size
plt.ylabel("Gas Price",fontdict= {'fontsize':20})

# Change tick marks on x axis
plt.xticks([1990,1992,1995,1997,2000,2002,2005,2007],["1990","1992","1995","1997","2000","2002","2005","2007"])

#Show Plot
plt.show()
```

# Gas Price Growth of Highest Prices 

```python
# Create plot
df1.plot(x="Year", y=["UK", "Germany"], kind="bar", figsize=(9, 8))

#Choose Title and custom font size 
plt.title("UK & Germany Gas Price Growth", fontdict= {'fontsize':25})

# Choose x lable and font size 
plt.xlabel("Year",fontdict= {'fontsize':20})

#Choose y label and custom font size 
plt.ylabel("Gas Price",fontdict= {'fontsize':20})

# Customize Legend 
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1,fontsize = 10)

#Show Graph
plt.show()
```

# Growth of Lowest Gas Price Countries

```python
# Plot Bar Graph
df1.plot(x="Year", y=["USA", "Mexico"], kind="bar", figsize=(9, 8),color = ["Purple", "Orange"])

#Custom Title and font size 
plt.title("USA & Mexico Gas Price Growth",fontdict= {'fontsize':25})

#Xlabel and custom font size 
plt.xlabel("Year",fontdict= {'fontsize':20})

#ylabel and font size 
plt.ylabel("Gas Price",fontdict= {'fontsize':20})

#Customize Legend
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1,fontsize = 10)

#Show Graph
plt.show()
```
